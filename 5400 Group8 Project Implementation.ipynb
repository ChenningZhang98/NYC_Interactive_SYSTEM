###  Step 3: Deploy a REST API web service using Python Flask
We implemented three interfaces with three HTML files; each is connected to a different route for execution. 

#### Template 1: Inspection Query Form.html
##### @app.route('/')
This is the first interface shown to the users. It includes an input form asking for restaurant's name, boro (neighborhood), and zipcode. Users can enter the information correspondingly. If the inputs are not valid, an error code of "No Match Records" will be shown.  

#### Template 2: results2.html
##### @app.route('/inspection', methods=['POST'])
After users enter the information of their restaurants and click the send button, our program will run the restult2.html form and return a table of inspection records that match users' inputs. At the bottom of the page, there will be a form of inspection comments. Users can identify the inspection record to dispute, fill out the form with the record's unique ID along with users' comments. If the input ID is not valid or out of range, an error code of "ID Not Found" will be shown.  

#### Template 3: comment.html
##### @app.route('/comment', methods=['POST'])
After users submit their comments, their comments will be inserted as a new key-value pair to corresponding inspection record in the original MongoDB database collection. Everytime users enter a new comment related to the record, the comment value will be automatically updated. 

```python
import pandas as pd
```

```python
from flask import Flask, request, render_template
app = Flask(__name__, template_folder = 'templates')

@app.route('/')

def my_form():
    return render_template("Inspection Query Form.html")

@app.route('/inspection', methods=['POST'])

def my_form_post():
    val_name= request.form['name']
    val_boro = request.form['boro']
    val_zipcode = request.form['zipcode']
    query_results = []
   
    query_results = collection.find({"DBA": val_name, "BORO": val_boro,"ZIPCODE":int(val_zipcode)})
    
    a = list(query_results)

    keys = ['ID','VIOLATION CODE','VIOLATION DESCRIPTION', 'INSPECTION TYPE','SCORE', 'GRADE DATE']
    
    df = pd.DataFrame(columns=keys)
    for i in range(0,len(a)):
        app = list(map(a[i].get, keys))
        dl = len(df)
        df.loc[dl] = app
   
    error_code = "No Match Inspection Records"
    if df.empty:
        return render_template('Inspection Query Form.html', val=error_code)
    else:
        return render_template('results2.html',  tables=[df.to_html(classes='data')], titles=df.columns.values)  


@app.route('/comment', methods=['POST'])

def my_form_retrieve():
    val_id = int(request.form['id'])
    val_comm = request.form['comment']
    
    query_comment = []
    collection.update_one({"ID": val_id},{"$set":{"Review":val_comm}})
    query_comment = collection.find({"ID": val_id})
    a = list(query_comment)
    keys = ['ID','VIOLATION CODE','VIOLATION DESCRIPTION', 'INSPECTION TYPE','SCORE', 'GRADE DATE',"Review"]
    df = pd.DataFrame(columns=keys)
    for i in range(0,len(a)):
        app = list(map(a[i].get, keys))
        dl = len(df)
        df.loc[dl] = app

    error_code = "ID Not Found"
    if df.empty:
        return render_template('results2.html', val=error_code)
    else:
        return render_template('comment.html',  tables=[df.to_html(classes='data')], titles=df.columns.values) 

app.run(host='localhost', port=5004)
```
